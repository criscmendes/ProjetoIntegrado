```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
ireplies = pd.read_csv('NewsItemsReplies.csv', low_memory=False)
ireplies.head()
```

```python
ireplies.drop_duplicates(inplace=True)
```

```python
## DROP_DUPLICATES CONDICIONADO ##
# Ordena IN_REPLY_TO_NEWS_ID de forma a que estejam primeiro os que estão 
# em notação cientifica e, por fim, elimina os duplicados mantendo só os 
# últimos que terão o formato normal caso existam.

ireplies.sort_values('in_reply_to_news_id',inplace=True)
ireplies.drop_duplicates(['comment_id','news_tweet_text','comment_tweet_date',
                          'comment_tweet_text'], keep='last', inplace=True)
ireplies
```

```python
import string
from nltk.corpus import stopwords
```

```python
ireplies=ireplies.sort_index()
ireplies
```

```python
def remove_arroba(mess):
    remove = re.sub('@([^\s]+)[ ]*', '', mess)
    return remove
```

```python
ireplies['comment_tweet_text'] = ireplies['comment_tweet_text'].apply(lambda x: remove_arroba(x))
ireplies.head()
```

```python
def remove_emoji(mess):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', mess)
```

```python
ireplies['comment_tweet_text'] = ireplies['comment_tweet_text'].apply(lambda x: remove_emoji(x))
ireplies.head()
```

```python
def remove_ponctuation(mess):
    
    #Retira pontuações
    nopunc = [char for char in mess if char not in string.punctuation]

    # Junta-os para formar strings
    nopunc = ''.join(nopunc)
    
    # Remove as stopwords
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

# Apagar links dos comentários

```python
def remove_links(linha):
    URLless = re.sub(r"http\S+", "", str(linha))
    return URLless
```

```python
ireplies['comment_tweet_text'] = ireplies['comment_tweet_text'].apply(lambda x: remove_links(x))
ireplies.head(50)
```
