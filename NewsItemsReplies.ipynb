```python
## DROP_DUPLICATES CONDICIONADO ##
# Ordena IN_REPLY_TO_NEWS_ID de forma a que estejam primeiro os que estão 
# em notação cientifica e, por fim, elimina os duplicados mantendo só os 
# últimos que terão o formato normal, caso existam.

ireplies.sort_values('in_reply_to_news_id',inplace=True)
ireplies.drop_duplicates(['comment_id','news_tweet_text','comment_tweet_date',
                          'comment_tweet_text'], keep='last', inplace=True)
ireplies
```

```python
# 1239325 (drop_duplicates(), low_memory=False)
# 1214909 [-24416] (drop_duplicates())
# 1114396 [-100513] (drop_duplicates('comment_id') condicionado, drop_duplicates(), low_memory=False)
ireplies.info()
```

### 'comment_id'

```python
# linhas com COMMENT_ID em notação cientifica
commID=ireplies[ireplies[['comment_id']]
         .apply(lambda x: x.str.contains(',|E|e', regex=True))
         .any(axis=1)]
commID.head()
```

```python
# 32 (low_memory=False and drop_duplicates())
# 28 (drop_duplicates(comment id, ..) com condição)
len(commID)
```

```python
# DROP das linhas presentes em commID [notação cientifica em 'comment_id']
ireplies.drop(commID.index,inplace=True)
```

### 'in_reply_to_news_id'

```python
# linhas com IN_REPLY_TO_NEWS_ID em notação cientifica
iRep2News = ireplies[ireplies[['in_reply_to_news_id']]
             .apply(lambda x: x.str.contains(',|E|e', regex=True))
             .any(axis=1)]
iRep2News.head()
```

```python
# 131961 (drop_duplicates(), low_memory=False)
# 98332 (drop_duplicates(), low_memory=False, drop_duplicates(comment_id, ...) condicionado)
# 32 (drop_duplicates(), low_memory=True)
len(iRep2News)
```

```python
# DROP das linhas presentes em iRep2News [notação científica em 'in_reply_to_news_id']
ireplies.drop(iRep2News.index,inplace=True)
```

### 'comment_id' e 'in_reply_to_news_id' [object => int 64]

```python
# Alteração de TIPOS das colunas 'comment_it' e 'in_reply_to_news_id' object->int64
ireplies = ireplies.astype({"comment_id": int, "in_reply_to_news_id": int})
ireplies.info()
```
